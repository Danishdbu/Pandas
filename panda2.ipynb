# how to rename cols  

```python
df = pd.DataFrame({'A_col':[1,2,3,4,5,6,7],'B_col':[11,44,33,22,77,88,33]})
df
```

```python
# rename cols 
df.rename(columns={"A_col": "a", "B_col": "c"},inplace=True)
```

```python
df
```

```python
# rename index
df.rename(index={0: "x", 1: "y", 2: "z",3:'s',4:'r',5:'t',6:'d'},inplace=True)
df
```

```python
df.columns=['col1','col2']
df
```
