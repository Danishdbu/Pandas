# Join

### types of join
`'left'
'right'
'outer'
'inner'`

### inner join

```python
students.merge(reg,how='inner',on='student_id')
```

#### left join

```python
courses.merge(reg,how='left',on='course_id')
```

```python
temp_df = pd.DataFrame({'student_id':[26,27,28],
              'name':['Nitish','Ankit','Rahul'],
              'partner':[28,26,17]})
temp_df 
```

```python
students = pd.concat([students,temp_df],ignore_index=True)
students
```

### right join

```python
students.merge(reg,how='right',on='student_id')
```

### outer join

```python
students.merge(reg,how='outer',on='student_id').tail(10)
```

```python
# 1. find total revenue generated
courses.merge(reg,how='inner',on='course_id')['price'].sum()
```

```python
f = courses.merge(reg,how='inner',on='course_id')
g = f.groupby('course_name').sum()
g['price'].sum()
```

```python
# 2. find month by month revenue 
temp_1 = pd.concat([nov,dec],keys=['Nov','Dec']).reset_index()
temp_1.merge(courses,how='inner',on='course_id').groupby('level_0')['price'].sum()
```

```python
# 3. Print the registration table
# cols -> name -> course -> price
print(reg.columns)
print(courses.columns)
print(students.columns)

```

```python
new_tab= reg.merge(courses,on='course_id').merge(students,on='student_id')[['name', 'course_name','price']]
new_tab
```

```python
new_tab.groupby(['name','course_name']).sum()
```

```python
# 4. Plot bar chart for revenue/course
reg.merge(courses,on='course_id').groupby('course_name')['price'].sum().plot(kind='bar')

```
