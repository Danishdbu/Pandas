# MERGE 

```python
x = pd.DataFrame({"A":[1,2,3,4,5],"B":[23,24,25,22,56]})
y = pd.DataFrame({"A":[1,2,3,4,5],"k":[3,24,2,22,5]})
pd.merge(x,y)
```

```python
x = pd.DataFrame({"A":[1,2,3,4,5],"B":[23,24,25,22,56]})
y = pd.DataFrame({"A":[1,2,3,4,5],"k":[3,24,2,22,5]})
pd.merge(x,y,on="A")
```

```python
x = pd.DataFrame({"A":[1,2,3,4,5],"B":[23,24,25,22,56]})
y = pd.DataFrame({"A":[1,2,3,4,6],"k":[3,24,2,22,5]})
pd.merge(x,y,how="left")
```

```python
x = pd.DataFrame({"A":[1,2,3,4,5],"B":[23,24,25,22,56]})
y = pd.DataFrame({"A":[1,2,3,4,6],"k":[3,24,2,22,5]})
pd.merge(x,y,how="right")
```

```python
df = {'age':[10,22,13,21,12,11,17],
      'section':['A','B','C','B','B','A','A'],
      'city':['Bangalore','Delhi', 'Mumbai', 'Delhi','Chandigarh','Mumbai','sultanpur'],
      'gender':['M','F','F','M','M','M','F'],
      'color':['red',None,'yellow','green',None,'red','black']}
```

```python
data = pd.DataFrame(df)
data
```

```python
data.loc[data.age >= 15]
```

```python
# select with multiple conditions
data.loc[(data.age >= 12) & (data.gender == 'M')]
```

```python
# select few columns with a condition
data.loc[(data.age >= 12), ['city', 'gender']]
```

```python
# update a column with condition
data.loc[(data.age >= 12), ['section']] = 'M'
data
```

```python
data.loc[(data.age >= 20), ['section', 'city']] = ['S','Pune']
data
```
